```python
import datetime
import amipy
from quant_iq import MsShagwell


class Context(object):
    ''' backtest context '''
    def __init__(self):
        self.symbol = 'FT.60min'  #*** FTSE 100 ICE FUTURE continuous ***#
        self.dbase = 'FUTURE'
        self.starting_equity = 100000.00
        self.margin_required = 5500.0
        self.tick_size = 0.5
        self.tick_value = 5.0
        self.risk = 0.15
        self.warmup_bars = 500
        self.commission = 4.50
        

START_DATE = datetime.datetime(2003, 7, 1)
END_DATE = datetime.datetime(2016, 12, 31)
```

```python
OBJ = Context()
```

```python
STRAT = MsShagwell(OBJ) #*** It's getting hot in here ;-) ***#
```

```python
OHLC = amipy.mongo_grab(OBJ.symbol, OBJ.dbase, START_DATE, END_DATE)
```

```python
STRAT.Run(OHLC)
```
