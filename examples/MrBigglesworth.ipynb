```python
import datetime
import amipy
from quant_iq import MrBigglesworth
```

```python
class Context(object):
    ''' backtest context '''
    def __init__(self):
        self.symbol = 'AMZN.30min'  #*** AMZN ***#
        self.dbase = 'SP100'
        self.starting_equity = 100000.00
        self.margin_required = 0 #*** use 0 for stocks ***#
        self.tick_size = 0.01
        self.tick_value = 0.01
        self.risk = 0.65
        self.warmup_bars = 500
        self.commission = 0.025 #*** per lot ***#
```

```python
START_DATE = datetime.datetime(2007, 1, 2)
END_DATE = datetime.datetime(2016, 12, 31)
```

```python
OBJ = Context()
```

```python
OHLC = amipy.mongo_grab(OBJ.symbol, OBJ.dbase, START_DATE, END_DATE)
```

```python
STRAT = MrBigglesworth(OBJ) #*** Mua hah ha! ***#
```

```python
STRAT.Run(OHLC)
```
