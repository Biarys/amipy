```python
"""
Created on Mon Jun  5 07:53:15 2017
@author: github.com/Quantmatic
"""
import datetime
import TTR
import amipy
from strategy_wyckoff_varadi import VolatilityAdjustedStrategy
```

```python
START_DATE = datetime.datetime(2006, 1, 2)
END_DATE = datetime.datetime(2016, 12, 31)
```

```python
INSTRUMENTS = [{'SYMBOL': '@ES#1H', 'DBASE': 'FUTURE', #*** e-mini sp500 60min continuous ***#
                'EQUITY': 100000, 'MARGIN': 5500,
                'TICKSIZE': 0.25, 'TICKVALUE': 12.5,
                'RISK': 0.07, 'WARMUP_BARS': 500},
               {'SYMBOL': '@NQ#1H', 'DBASE': 'FUTURE', #*** e-mini Nasdaq 60min continuous ***#
                'EQUITY': 100000, 'MARGIN': 5500,
                'TICKSIZE': 0.25, 'TICKVALUE': 5.0,
                'RISK': 0.07, 'WARMUP_BARS': 500},
               {'SYMBOL': '@TY#1H', 'DBASE': 'FUTURE', #*** 10-Year Note 60min continuous ***#
                'EQUITY': 200000, 'MARGIN': 5500,
                'TICKSIZE': 0.005, 'TICKVALUE': 15.625,
                'RISK': 0.04, 'WARMUP_BARS': 500}]
```

```python
for inst in INSTRUMENTS:
    context = type('context', (object, ), {'symbol': '', 'dbase': ''})
    context.symbol = inst['SYMBOL']
    context.dbase = inst['DBASE']
    context.starting_equity = inst['EQUITY']
    context.margin_required = inst['MARGIN']
    context.tick_size = inst['TICKSIZE']
    context.tick_value = inst['TICKVALUE']
    context.risk = inst['RISK']
    context.warmup_bars = inst['WARMUP_BARS']
    context.commission = 4.50

    OHLC = amipy.mongo_grab(context.symbol, context.dbase, START_DATE, END_DATE)

    STRAT = VolatilityAdjustedStrategy(context)

    STRAT.Run(OHLC)
```
