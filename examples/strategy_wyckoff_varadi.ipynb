```python
import datetime
import amipy
from strategy_wyckoff_varadi import VolatilityAdjustedStrategy

class Context(object):
    ''' backtest context '''
    def __init__(self):
        self.symbol = '@ES#1H'  #*** e-mini SP 60min continuous ***#
        self.dbase = 'FUTURE'
        self.starting_equity = 100000.00
        self.margin_required = 5500.0
        self.tick_size = 0.25
        self.tick_value = 12.5
        self.risk = 0.15
        self.warmup_bars = 500
        self.commission = 4.50
        
START_DATE = datetime.datetime(2006, 1, 2)
END_DATE = datetime.datetime(2016, 12, 31)
```

```python
OBJ = Context()

STRAT = VolatilityAdjustedStrategy(OBJ)

OHLC = amipy.mongo_grab(OBJ.symbol, OBJ.dbase, START_DATE, END_DATE)
```

```python
STRAT.Run(OHLC)
```
